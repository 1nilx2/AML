# Section A

```python
import pandas as pd
df_tmp = pd.read_csv('HW1_result.csv')
df_tmp.head()
```

```python
import nltk
from nltk import word_tokenize

#Make an empty list to store each comment.
tmp = ['']
series_comments = df_tmp['comments']

#From the comments Series, extract all the corpus and then make them in a row.
for i in range (len(series_comments)):
    tmp[0] = tmp[0] + str(series_comments[i]).lower()
    
#Then Tokenize the corpus into tokens
token = nltk.word_tokenize(tmp[0])
```

```python
#From the tokenized words, the word frequency can be easily caculated wth freqdist function.
fd = nltk.FreqDist(token)

#From the fact that the FreqDist has shape like dictionary, 
#I could easily convert the FreqDist into a dataframe which can be more versatile in my opinion.
df_fdist = pd.DataFrame.from_dict(fd, orient='index')

#These two commands are not required, but exist just for being nice.
df_fdist.columns = ['Frequency']
df_fdist.index.name = 'Word'

#Let's sort the value by the frequency as i need to check zipf's law
df_fdist.sort_values(by='Frequency', ascending = False, inplace=True)

#This is the Frequency Data frame including punctuations. 
df_fdist[:20]
```

## Let's remove some punctuation which might blur our eyes

```python
#From now, I will conduct similiar analysis, but without punctuations.
df_tmp_npunc = pd.read_csv('HW1_result.csv')
df_tmp_npunc.head()

df_tmp_npunc["new_column"] = df_tmp_npunc['comments'].str.replace('[^\w\s]','')

#Make an empty list to store each comment.
tmp_npunc = ['']
series_comments_npunc = df_tmp_npunc['new_column']

#From the comments Series, extract all the corpus and then make them in a row.
for i in range (len(series_comments_npunc)):
    tmp_npunc[0] = tmp_npunc[0] + str(series_comments_npunc[i]).lower()
    
#Then Tokenize the corpus into tokens
token_npunc = nltk.word_tokenize(tmp_npunc[0])
```

```python
#From the tokenized words, the word frequency can be easily caculated wth freqdist function.
fd_npunc = nltk.FreqDist(token_npunc)

#From the fact that the FreqDist has shape like dictionary, 
#I could easily convert the FreqDist into a dataframe which can be more versatile in my opinion.
df_fdist_npunc = pd.DataFrame.from_dict(fd_npunc, orient='index')

#These two commands are not required, but exist just for being nice.
df_fdist_npunc.columns = ['Frequency']
df_fdist_npunc.index.name = 'Word'

#Let's sort the value by the frequency as i need to check zipf's law
df_fdist_npunc.sort_values(by='Frequency', ascending = False, inplace=True)

#This is the Frequency Data frame including punctuations. 
df_fdist_npunc[:20]
```

## Let's make an dataframe illustrating the zipf's law
### Set the initial value of the zipf's law as same with the frequency of the most frequent word

```python
#I will createa dataframe from a list. 
#The length of the list will be 100 as I will match this with top-100 frequent words
#Make the starting value same with the Frequency so we don't need to worry about the scale
list_tmp = [df_fdist_npunc.iloc[0]['Frequency']]
list_tmp

for i in range(1,100):
    list_tmp.append(0)

df_zipf = pd.DataFrame(list_tmp)
df_zipf.rename(columns = {0:"zipf's value"}, inplace=True)
df_zipf

for i in range(1,100):
    df_zipf.iloc[i]["zipf's value"] = df_zipf.iloc[0]["zipf's value"] * (1/(i+1))
    
df_zipf
```

```python
#Make the list of the indices in order to easily plot the graph
index_list = list(df_fdist_npunc.index.values)
index_zipf = list(df_zipf.index.values)

for i in range(100):
    index_zipf[i] +=1 
```

```python
import matplotlib.pyplot as plt
    
plt.figure(figsize=(20,6))
plt.subplot(121)
ax = plt.gca()

#Let's draw the graph related to the top 100 words
ax.plot(index_list[:100], df_fdist_npunc['Frequency'][:100])
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title("Frequency of Words excepting punctuations")
plt.axis('tight')

#For zipf's law df
    
plt.subplot(122)
ax = plt.gca()
ax.plot(index_zipf[:100], df_zipf["zipf's value"][:100])
plt.xlabel('Ranking')
plt.ylabel("zipf's value")
plt.title("zipf's law")
plt.axis('tight')

plt.show()
```

# Section B

```python
from nltk.corpus import stopwords

#Setting the stopwords set which will play a role to delete them from the tokenized words.
stop_words = set(stopwords.words('english'))

#Bring the car-model information, by which the words indicating model
#will be transformed into the matched brand
df_car = pd.read_csv('car models and brands.csv')
df_car.set_index(df_car['Model'], inplace=True)
df_car = df_car[['Brand']]
df_car
```

```python
#set the new df
df_b = pd.read_csv('HW1_result.csv')
df_b["new_column"] = df_b['comments'].str.replace('[^\w\s]','')
df_b['Temp2'] = df_b["new_column"].replace(to_replace=df_car.index, value=df_car['Brand'], regex=True)

#Make an empty list to store each comment.
tmp_xstop = ['']
series_comments_xstop = df_b['Temp2']
series_comments_xstop

#From the comments Series, extract all the corpus and then make them in a row.
for i in range (len(series_comments_xstop)):
    tmp_xstop[0] = tmp_xstop[0] + str(series_comments_xstop[i]).lower()
        
# #Then Tokenize the corpus into tokens
token_xstop = nltk.word_tokenize(tmp_xstop[0])
```

```python
#Filter the stop_words
filtered_xstop = [w for w in token_xstop if w not in stop_words]

#From the tokenized words, the word frequency can be easily caculated wth freqdist function.
fd_xstop = nltk.FreqDist(filtered_xstop)

#From the fact that the FreqDist has shape like dictionary, 
#I could easily convert the FreqDist into a dataframe which can be more versatile in my opinion.
df_xstop = pd.DataFrame.from_dict(fd_xstop, orient='index')

#These two commands are not required, but exist just for being nice.
df_xstop.columns = ['Frequency']
df_xstop.index.name = 'Word'

#Let's sort the value by the frequency as i need to check zipf's law
df_xstop.sort_values(by='Frequency', ascending = False, inplace=True)

#This is the Frequency Data frame including punctuations. 
df_xstop[40:60]
```
